# Module 1: ETL with PySpark

```python
# Run this cell first
!pip install -q pyspark

from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *

spark = SparkSession.builder.appName("LearningModules").getOrCreate()
sc = spark.sparkContext
```

```python
# Load datasets (adjust path if needed)
students_df = spark.read.csv("../data/students.csv", header=True, inferSchema=True)
enrollments_df = spark.read.csv("../data/enrollments.csv", header=True, inferSchema=True)
```

## Extract: Reading Data

The first step in ETL is extracting data from a source. PySpark can read CSV, JSON, Parquet, and many other formats.

```python
# Show the schema of our DataFrame
students_df.printSchema()
```

```python
# Show the first 5 rows
students_df.show(5)
```

## Transform: Selecting and Filtering

Transformations reshape, filter, and enrich the data.

```python
# Select specific columns
selected_df = students_df.select("student_id", "name", "gpa")
selected_df.show(5)
```

```python
# Filter students with GPA > 3.5
high_gpa_df = students_df.filter(col("gpa") > 3.5)
high_gpa_df.show()
```

## Practice Problem 3: Add a Derived Column

Add a column called `gpa_category` that labels students as "High" if GPA >= 3.5, otherwise "Standard".

<details><summary>Hint</summary>Use withColumn() with when().otherwise().</details>

```python
# Your code here
```

<details><summary>Solution</summary></details>

```python
# Solution
students_df.withColumn(
    "gpa_category",
    when(col("gpa") >= 3.5, "High").otherwise("Standard")
).show()
```
