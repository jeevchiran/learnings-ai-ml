```python
# Run this cell first
!pip install -q pyspark

from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *

spark = SparkSession.builder.appName("LearningModules").getOrCreate()
sc = spark.sparkContext
```

```python
# Load datasets (adjust path if needed)
students_df = spark.read.csv("../data/students.csv", header=True, inferSchema=True)
enrollments_df = spark.read.csv("../data/enrollments.csv", header=True, inferSchema=True)
```

## Schema and Structure

Start exploration by understanding the structure of your data.

```python
# Print the schema
students_df.printSchema()

# Show column names and count
print(f"Columns: {students_df.columns}")
print(f"Number of rows: {students_df.count()}")
print(f"Number of columns: {len(students_df.columns)}")
```

## Statistical Summary with describe() and summary()

```python
# describe() gives count, mean, stddev, min, max
students_df.describe().show()
```

```python
# summary() gives additional percentiles
students_df.summary().show()
```

## Group Counts and Distributions

```python
# Count by group
enrollments_df.groupBy("course_id").count().orderBy(desc("count")).show()
```

```python
# Distinct count
print(f"Distinct students: {students_df.select('student_id').distinct().count()}")
print(f"Distinct courses: {enrollments_df.select('course_id').distinct().count()}")
```

## Crosstab

Crosstab shows the frequency distribution of two columns.

```python
# Add a GPA category for crosstab
students_with_category = students_df.withColumn(
    "gpa_level",
    when(col("gpa") >= 3.5, "High")
    .when(col("gpa") >= 3.0, "Medium")
    .otherwise("Low")
)

# Join with enrollments and crosstab
joined = students_with_category.join(enrollments_df, "student_id")
joined.crosstab("gpa_level", "course_id").show()
```

## Practice Problem 3: Multi-Column Summary
