```python
# Run this cell first
!pip install -q pyspark

from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *

spark = SparkSession.builder.appName("LearningModules").getOrCreate()
sc = spark.sparkContext
```

```python
# Load datasets (adjust path if needed)
students_df = spark.read.csv("../data/students.csv", header=True, inferSchema=True)
enrollments_df = spark.read.csv("../data/enrollments.csv", header=True, inferSchema=True)
```

```python
import pandas as pd

# Create equivalent Pandas DataFrames
students_pdf = students_df.toPandas()
enrollments_pdf = enrollments_df.toPandas()
```

## Select vs [] (Column Selection)

Both frameworks let you pick specific columns, but with different syntax.

```python
# PySpark: select
students_df.select("name", "gpa").show(5)

# Pandas: bracket notation
print(students_pdf[["name", "gpa"]].head())
```

## Filter vs query (Row Filtering)

```python
# PySpark: filter
students_df.filter(col("gpa") > 3.5).show(5)

# Pandas: query or boolean indexing
print(students_pdf.query("gpa > 3.5").head())
# or: students_pdf[students_pdf["gpa"] > 3.5]
```

## groupBy vs groupby (Aggregation)

```python
# PySpark: groupBy
students_df.groupBy("gpa").count().show(5)

# Pandas: groupby
print(students_pdf.groupby("gpa").size().head())
```

## withColumn vs Direct Assignment (Adding Columns)

```python
# PySpark: withColumn (returns a new DataFrame)
students_df.withColumn("gpa_x2", col("gpa") * 2).show(5)

# Pandas: direct assignment (modifies in place or creates copy)
students_pdf_copy = students_pdf.copy()
students_pdf_copy["gpa_x2"] = students_pdf_copy["gpa"] * 2
print(students_pdf_copy.head())
```

## Key Differences Summary

| Operation | PySpark | Pandas |
|-----------|---------|--------|
| Select columns | `.select("col")` | `df[["col"]]` |
| Filter rows | `.filter(condition)` | `.query()` or boolean index |
| Group & Aggregate | `.groupBy().agg()` | `.groupby().agg()` |
| Add column | `.withColumn("new", expr)` | `df["new"] = expr` |
| Immutable | Yes (new DF each time) | No (in-place possible) |
| Execution | Lazy | Eager |

## Practice Problem 4: Add a Computed Column

Add a column `pass_fail` that is "Pass" if GPA >= 2.0 else "Fail". Do this in both frameworks.

<details><summary>Hint</summary>PySpark: withColumn + when/otherwise. Pandas: np.where or apply.</details>

```python
# Your code here
```

<details><summary>Solution</summary></details>

```python
# Solution
import numpy as np

# PySpark
students_df.withColumn(
    "pass_fail", when(col("gpa") >= 2.0, "Pass").otherwise("Fail")
).show(5)

# Pandas
students_pdf_copy = students_pdf.copy()
students_pdf_copy["pass_fail"] = np.where(students_pdf_copy["gpa"] >= 2.0, "Pass", "Fail")
print(students_pdf_copy.head())
```
